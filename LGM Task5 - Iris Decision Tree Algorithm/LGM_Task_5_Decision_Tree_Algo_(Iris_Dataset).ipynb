## Importing Required *Modules*

```python
# Basic Data Exploration and Visualization Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data Preprocessing and Modelling Libraries

from sklearn.preprocessing import LabelEncoder , StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree , export_text
# Error measurement methods

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , ConfusionMatrixDisplay
```

```python
data = pd.read_csv('/content/Iris.csv' , index_col = 'Id')
data.head()
```

```python
data.shape
```

```python
data.size
```

```python
data.info()
```

```python
data.describe()
```

```python
le = LabelEncoder()

data['Species'] = le.fit_transform(data['Species'])

data.head()
     
```

```python
sns.set_style('white')

sns.pairplot(data , hue="Species")


plt.show()
```

```python
sns.scatterplot(x=data['PetalLengthCm'] , y=data['PetalWidthCm'] , hue=data['Species'] , marker='*')
plt.xlabel("Petal Lenght")
plt.ylabel("Petal Widht")
plt.show()
```

```python
sns.scatterplot(x=data['SepalLengthCm'] , y=data['SepalWidthCm'] , hue=data['Species'] , marker='+')
plt.xlabel("Setal Lenght")
plt.ylabel("Setal Widht")
plt.show()
```

```python
X = data.drop('Species' , axis = 1)
y = data['Species']
     
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.25 , random_state = 50)
```

## ***Decision Tree Algorithm***

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
tree.score(X_test,y_test)
```

```python
y_pred = tree.predict(X_test)
```

```python
print("The accuracy of Decision Tree is :",accuracy_score(y_test,y_pred)*100)
```

```python
confusion_mat = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(confusion_mat).plot()
plt.show()
```

```python
plt.figure(figsize=(25,12))
feature = ['Petal Lenght (cm)','Petal Width (cm)','Setal Lenght (cm)','Setal Width (cm)']
target = ['Species']
plot_tree(tree ,feature_names='feature' , class_names='target' , filled = True)
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
